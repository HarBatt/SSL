```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/mpro_exp_data2_rdkit_feat.csv", index_col=0)
df.head()
```

```python
df['feat_10'].describe()
```

```python
cols = ["lib_name", "cmpd_id", "smiles", "label", "subset"]
label = df["label"]
subset = df["subset"]
new_df = df.drop(columns = cols)
```

```python
new_df.head()
```

```python
for col in new_df.columns:
    if len(new_df[col].unique()) == 1:
        new_df.drop(col,inplace=True,axis=1)
```

```python
new_df["feat_205"].describe()
```

```python
new_df = new_df.replace([np.inf, -np.inf], np.nan)
```

```python
new_df["feat_205"].describe()
```

```python
new_df["feat_11"].isna().sum()
```

```python
new_df = new_df.fillna(new_df.mean())
```

```python
new_df["feat_11"].isna().sum()
```

```python
new_df['label'] = label
new_df['subset'] = subset
```

```python
new_df.to_csv('data/clean_data.csv')
```

```python
new_df.head()
```

```python
train = new_df[(new_df["subset"] == "train")]
valid = new_df[(new_df["subset"] == "valid")]
test = new_df[(new_df["subset"] == "test")]
```

```python
train.to_csv('data/train_data.csv')
```

```python
valid.to_csv('data/valid_data.csv')
```

```python
test.to_csv('data/test_data.csv')
```
